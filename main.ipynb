```python
from mpl_toolkits import mplot3d
def f(x, y):
    return 9*x**2 + 9*y**3

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')




```

```python
NUM_POINTS = 40
points = df.close.values[-NUM_POINTS:]
if points is None:
    print('No data')
    exit(1)
extremums = []
plt.plot(points, 'bo')
plt.plot(points)

for i in range(1, NUM_POINTS-1):
    if points[i-1] < points[i] > points[i+1] or points[i-1] > points[i] < points[i+1]:
        extremums.append(i)
        plt.plot(i, points[i], 'ro')
```
